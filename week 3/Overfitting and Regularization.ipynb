# Part II: Addressing Overfitting

**Collect More Training Example**

- Having more training examples can make the model to generalize better
- Cons: we may not have enough data or cost-insufficient

**Feature Selection**

- Using all features: some may not have enough data, thus may leads to overfit
- Choosing only a subset of features that (intuitively) matters can better fit the data, but valuable features may be lost in the process, this is called feature selection

**Regularization**

- Instead of deselect a feature (by setting its weight to 0), we can minimize its weight instead of eliminating them
- There are algorithms for automatically do this
- Generally, it is better to regularize the weights, rather than the bias

# Part III: Cost Function of Regularization

**Intuition**

- Suppose we decided to use all parameters $w_{1}, w_{2}, w_{3}, w_{4}$, and regularized $w_{3}, w_{4}$ such that they are closed to 0.
- Then for the cost function, we can add these two variables, multiplied by a (arbitrarily) large scalar, to penalize for the difference

**What Regularization Do**

- We would choose to consider all 100 features and add a regularization term, lambda, to the cost function
- We may include b, but in practice, this would make little effect on the model

**Regularization**

- If lambda is 0, we are not penalize our parameters for overfitting, the extra terms would simply be 0
- If lambda is large, all our parameters are likely 0, then this would fit a straightline of $f_{\vec{w}, b}(\vec{x}) = b$
- Therefore, we need to choose lambda such that this balances out
