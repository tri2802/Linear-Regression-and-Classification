# Part II: Logistic Regression

**The Sigmoid Function**


- The sigmoid function outputs values 0 or 1
- We use g(z) to denote this function: $$g(z) = \frac{1}{1 + e^{-z}}$$  where 0 < g(z) < 1

- Notice if z is large, then $e^{-z}$ is small and g(z) = 1
- Conversely, if z is small, then g(z) = 0

- Recall the linear regression function, let z be: $$z = f_{\vec{w}, b}(\vec{x}) = \vec{w}\cdot\vec{x} + b$$  
then $$g(z) = \frac{1}{1 + e^{-(\vec{w}\cdot\vec{x} + b)}}$$

- The logistic regression function is denoted as:  
    $$f_{\vec{w}, b}(\vec{x}) = g(\vec{w}\cdot\vec{x} + b)) = \frac{1}{1 + e^{-(\vec{w}\cdot\vec{x} + b)}}$$

**Interpretation of Logistic Regression output**

$$f_{\vec{w}, b}(\vec{x})  = \frac{1}{1 + e^{-(\vec{w}\cdot\vec{x} + b)}}$$  
The probability of the ouput is 1

- Example:
x is tumor size
y is 0 (not malignant)
     1 (malignant)
- $f_{\vec{w}, b}(\vec{x}) = 0.7$ meaning 70% chance y is 1 

# Part III: Decision Boundary

- In this case there are 2 variables, so: $$f_{\vec{w}, b}(\vec{x}) = g(z) = g(w_{1}x_{1} + w_{2}x_{2} + b)$$  
- Decision boundary: z = $\vec{w} \cdot \vec{x} + b$ = 0  
Let $w_{1} = w_{2} = 1, b = -3$, then z = 0 meaning $x_{1} + x_{2} = 3$

**Non-linear Decision Boundary**


$$f_{\vec{w}, b}(\vec{x}) = g(z) = g(w_{1} x_{1}^2 + w_{2} x_{2}^2 + b)$$  
Let $w_{1} = w_{2} = 1, b = -1$, then:  
z = $x_{1}^2 + x_{2}^2 -1 = 0$  
    = $x_{1}^2 + x_{2}^2 = 1$
