**Training Set**


- where:  
i = 1, ..., m training examples  
j = 1, ..., n features   
y = 0 or 1  
- How to choose $\vec{w} = [w_{1} ... w_{n}]$ and b

**Squared Error Cost**
$$J(\vec{w}, b) = \frac{1}{m} \sum_{i = 1}^{m} \frac{1}{2} (f_{\vec{w},{b}}(\vec{x^{(i)}}) - y^{(i)})^{2}$$ 

- Linear Regression: $$f_{\vec{w}, b}(\vec{x}) = \vec{w} \cdot \vec{x} + b$$  

Gradient descent will take step by step to convergence (global minimum)

- Logistic Regression: $$f_{\vec{w}, b}(\vec{x}) = \frac{1}{1 + e^{-(\vec{w} \cdot \vec{x} + b)}}$$  
  
Gradient Descent will produce a non-convex function, as there will be many local minimas

- Defining the Loss:  
Denote the term: $$L(f_{\vec{w}, b}(\vec{x^{(i)}}), y^{(i)}) = \frac{1}{2} (f_{\vec{w},{b}}(\vec{x^{(i)}}) - y^{(i)})^{2}$$  
So the loss equals one-half of the squared difference of prediction and actual

**Logistic Loss Function**


- If $y^{(i)} = 1$
  
As $f_{\vec{w}, b}(\vec{x})$ goes to 1 then loss goes to 0  
As $f_{\vec{w}, b}(\vec{x})$ goes to 0 then loss goes to $\infty$  
- Then loss is lowest when $f_{\vec{w}, b}(x^{(i)})$ predicts closed to the label $y^{(i)}$

- Conversely, if $y^{(i)} = 0$  
  
As $f_{\vec{w}, b}(\vec{x})$ goes to 0 then loss goes to 0  
As $f_{\vec{w}, b}(\vec{x})$ goes to 1 then loss goes to $\infty$
