# SUPERVISED LEARNING

**Supervised Learning:** refers to algorithms that learn from x to y, or input to output mappings. 
- The key characteristic of supervised learning is that you give your learning algorithm **examples** to learn from
- Giving them right answers, known as the correct label y for a given input x, and is by seeing correct pairs of input x and desired output label y, then learning algorithm eventually learn to give accurate prediction of y from x

**Example of Machine Learning:**

- if the data collected is email and we want to know if it is a spam
- in self-driving car, we can create an algorithm that takes in information such as image, radar, so that we can position of other cars
- more popularly, we can collect audio, and through speech recognition, we can generate transcripts

**Another example: Regression - Housing price prediction**

- if we have a data set of house sizes and price, we can apply regression to predict the prize given the size
- One way is to fit a straight line (called linear regression); in this case, for the house that is 750 feet squared, the algorithm predicts the price to be 150k.
- However, there are multiple models to predict data, such as fitting a curve, which would leverage the housing price to 200k, thus giving higher profit. 

__Regression:__ predicts a number from infinitely many possible outputs. 

# CLASSIFICATION

**Example: Breast Cancer Detection:** Using a patient's medical records your machine learning system tries to figure out if a tumor that is a lump is malignant or is benign

- Let 0 denotes benign and 1 denotes malignant, we plot tumor size on the x axis and the diagnosis on the y axis
- One reason this differs from regression is that we only try to predict a small number of outputs or categories (in this case 0 and 1), this is what makes this classification

- We can also plot the data likes this since there are only 2 possible outputs in this example.

- In classification, we can take more than 2 output categories. 

**Classification:** predicts categories, which does not have to be numbers (like 0 and 1). For example, it can predict whether a picture is a cat or a dog. 

**Example:** Two or more inputs

- The algorithm here can take in 2 inputs, age and tumor size
- We will plot data on whether a tumor is malignant or benign given these two factors. 
- The algorithm has to decide how to fit a boundary line that separates the output categories.
- Note that we can also have more than 2 inputs as well. 

# UNSUPERVISED MACHINE LEARNING

**Reminder of Supervised Machine Learning**

- Given a set of data/examples
- Use an algorithm to predict the right answers

**Unsupervised Learning**

- We are not asked to predict the right answer given the data
- Instead, we are asked to find pattern, structures, or something interesting about the dataset
- The algorithm decides that the data can be divided into two different groups/clusters
> this is a particular type of unsupervised learning, called **clustering algorithm**

**Example: Clustering - Google News**

- The headline article talks about a giant panda, and the sub-particles follow the same categories
- Note that there are also similar keywords found from the article lists, such as 'zoo' and 'twin'
- This shows that the household has conducted many searches regarding this topic
> The algorithm searches through millions of articles to look for those that mention these keywords
-> Group them into clusters
- The algorithm figures this out on its own, without any supervision from human

**Another example: Clustering - DNA Microarray**

- Each tiny column represents the genetic or DNA activity of a person
- Each row represents a special type of Gene
- The idea is to measure how much certain genes, are expressed for each individual person
- Therefore, these colors, green and red, shows the degree which a person does or does not have a gene active
> We can run a clustering algorithm to group individuals into different categories
- We basically said to the algorithm here's a bunch of data, somehow find a way to structure it

**Another example: Clustering - Grouping Customers**

- We are given a database of customer information
> Our purpose is to split customers, based on their activities, into groups to better advertise/serve them **automatically**
